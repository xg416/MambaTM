```python
def nollToZernInd(j):
    """
    This function maps the input "j" to the (row, column) of the Zernike pyramid using the Noll numbering scheme.
    Authors: Tim van Werkhoven, Jason Saredy
    See: https://github.com/tvwerkhoven/libtim-py/blob/master/libtim/zern.py
    """
    if (j == 0):
        raise ValueError("Noll indices start at 1, 0 is invalid.")
    n = 0
    j1 = j-1
    while (j1 > n):
        n += 1
        j1 -= n
    m = (-1)**j * ((n % 2) + 2 * int((j1+((n+1)%2)) / 2.0 ))

    return n, m

def xor(x, y):
    return bool((x and not y) or (not x and y))

def tak_yam_int(i, j, s_arr):
    arg1 = 2 * s_arr
    arg2 = 0*1e-30
    smax = int(np.amax(s_arr))
    
    h = tak_indicator(i, j)
    
    ni, mi = nollToZernInd(i)
    nj, mj = nollToZernInd(j)
    
    nplus, nminus, mplus, mminus = ni+nj, ni-nj, mi+mj, mi-mj
    int1, int2 = s_arr*0, s_arr*0

    if h == 1:
        int1, a = integrate.quad_vec(lambda x: jv(mplus, arg1*x) * jv(ni+1, x) * jv(nj+1, x) / (x * (x**2 + arg2**2)**(11/6)), 1E-20, np.inf, limit=10**5)
        int2, a = integrate.quad_vec(lambda x: jv(np.abs(mminus), arg1*x) * jv(ni+1, x) * jv(nj+1, x) / (x * (x**2 + arg2**2)**(11/6)), 1E-20, np.inf, limit=10**5)

    if h == 2:
        int1, a = integrate.quad_vec(lambda x: jv(mplus, arg1*x) * jv(ni+1, x) * jv(nj+1, x) / (x * (x**2 + arg2**2)**(11/6)), 1E-20, np.inf, limit=10**5)
        int2, a = integrate.quad_vec(lambda x: jv(np.abs(mminus), arg1*x) * jv(ni+1, x) * jv(nj+1, x) / (x * (x**2 + arg2**2)**(11/6)), 1E-20, np.inf, limit=10**5)

    if h == 3:
        int1, a = integrate.quad_vec(lambda x: jv(mplus, arg1*x) * jv(ni+1, x) * jv(nj+1, x) / (x * (x**2 + arg2**2)**(11/6)), 1E-20, np.inf, limit=10**5)

    if h == 4:
        int1, a = integrate.quad_vec(lambda x: jv(mplus, arg1*x) * jv(ni+1, x) * jv(nj+1, x) / (x * (x**2 + arg2**2)**(11/6)), 1E-20, np.inf, limit=10**5)

    if h == 5:
        int1, a = integrate.quad_vec(lambda x: jv(mplus, arg1*x) * jv(ni+1, x) * jv(nj+1, x) / (x * (x**2 + arg2**2)**(11/6)), 1E-20, np.inf, limit=10**5)
```
