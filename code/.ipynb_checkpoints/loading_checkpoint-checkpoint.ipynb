```python
import torch
import argparse
from TM_model import Model
from utils.scheduler import GradualWarmupScheduler
import torch.optim as optim
```

```python
def get_args():
    parser = argparse.ArgumentParser(description=' ')
    parser.add_argument('--model', type=str, default='MambaTM3', help='type of model to construct')
    parser.add_argument('--n_features', type=int, default=16, help='base # of channels for Conv')
    parser.add_argument('--n_blocks', type=int, default=6, help='# of blocks in middle part of the model')
    parser.add_argument('--version', type=str, default='v2', help='version')
    return parser.parse_args()
```

```python
model_path = "/home/zhan3275/data/log_rnn/Mamba/MambaTM3_10-21-2024-22-55-43/checkpoints/model_240000.pth"
```

```python
class params:
    # I think I need to include these to allow hinting in Komodo. I think.
    def __init__(self):
        self.n_features = 16
        self.n_blocks = 6
        self.version = "v2"
        self.model = "MambaTM3"
para = params()
```

```python
model = Model(para).cuda()
```

```python
model2 = Model(para).cuda()
```

```python
checkpoint = torch.load(model_path)
```

```python
checkpoint.keys()
```

```python
optimizer = optim.Adam(model.parameters(), lr=0.0002, betas=(0.9, 0.99), eps=1e-8)
scheduler_cosine = optim.lr_scheduler.CosineAnnealingWarmRestarts(optimizer, 400000-10000, eta_min=1e-6)
scheduler = GradualWarmupScheduler(optimizer, multiplier=1, total_epoch=10000, after_scheduler=scheduler_cosine)
```

```python
scheduler.after_scheduler.state_dict()
```
